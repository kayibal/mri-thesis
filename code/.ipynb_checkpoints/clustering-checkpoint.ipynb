```python
def plot_cluster(clusters):
    cluster_counter = Counter([x for x in clusters])
    labels, values = zip(*cluster_counter.items())
    indexes = np.arange(len(labels))
    width = 1
    plt.bar(indexes, values, width)
    #plt.xticks(indexes + width * 0.5, labels)
```

```python
def generate_cluster_dict(cluster,paths):
    cluster_dic = {}
    indizes = np.arange(len(paths))
    for cluster, path, idx in zip(cluster,paths,indizes):
        if cluster_dic.has_key(cluster):
            cluster_dic[cluster].append({'path' : path, 'idx': idx})
        else:
            cluster_dic[cluster] = [{'path' : path, 'idx': idx}]
    return cluster_dic
```

```python
def retrieve_liked_clusters(cluster,likes):
    liked_clusters = []
    query = "SELECT %s_cluster FROM tracks WHERE msd_id = '%s'"
    
    conn = sqlite3.connect("/home/kayibal/sc-recom/code/msd_data/analysis_data/sampled_likes_tracks.db")
    c = conn.cursor()
    for msd_id in likes:
        c.execute(query % (cluster,msd_id))
        liked_clusters.append(c.fetchone()[0])
    conn.close()
    return liked_clusters
```

```python
def get_feature_index(msd_id):
    for i,track in enumerate(tracks):
        if track == msd_id:
            return i
```

```python
def get_distance(A,B, feature, part='all'):
    if part == 'mfcc':
        return euclidean(data[feature]['data'][get_feature_index(A),60:],data[feature]['data'][get_feature_index(B),60:])
    if part =="fluc":
        return euclidean(data[feature]['data'][get_feature_index(A),:60],data[feature]['data'][get_feature_index(B),:60])
    else:
        return euclidean(data[feature]['data'][get_feature_index(A),:],data[feature]['data'][get_feature_index(B),:])
```

<h3>Getting originial Fluctuation Pattern statistics</h3>
<p>
The following function loads all original fluctuation patterns from disk and calculates a mean and a std deviation matrix for each entry.
</p>

```python
def get_cluster_fluc_stats(cluster_dic, cluster):
    flucs = np.zeros((24,60,len(cluster_dic[cluster])))
    for idx,entry in enumerate(cluster_dic[cluster]):
        flucs[:,:,idx]=(np.load(os.path.join("/home/kayibal/sc-recom/code/msd_data/analysis_data/",entry['path'].replace('.mp3','_fluc.npy'))))
    mean = np.mean(flucs,axis=2)
    std = np.std(flucs,axis=2)
    return mean, std
```

```python
def get_cluster_feature_stats(data, cluster_dic, cluster):
    features = np.zeros((data.shape[0], len(cluster_dic[cluster])))
    idx = [entry['idx'] for entry in cluster_dic[cluster]]
    return np.mean(data[idx], axis = 0), np.std(data[idx], axis=0)
    
```
